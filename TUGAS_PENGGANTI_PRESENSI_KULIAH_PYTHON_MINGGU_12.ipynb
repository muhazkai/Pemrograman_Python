```
import numpy as np
```

```
import matplotlib.pyplot as plt
```

```
def cos(fc,c,ht,hr,d):
  #fc = frekuensi (MHz)
  #c = urban, suburban, open
  #ht = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if fc in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif fc in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if c=='urban' :
        Cm=0
    elif c=='suburban' :
        Cm= -2*((np.log10((fc/28)**2))-5.4)
    elif c=='open' :
        Cm= -4.78*(np.log10(fc))+18.33*(np.log10(fc))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(fc)-0.7)*hr-(1.56*np.log10(fc)-1.8)
    Lp = Ca+(Cb*np.log10(fc))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
fc=700 #MHz
fc=900 #MHz
fc=1800 #MHz
loc='urban'
loc='suburban'
loc='open'
tower=80 #m
user=1.5 #m
dmin=200 #m
dmax=700 #m
N=200 #titik
d=np.random.randint(dmin,dmax,N)
```

```
print(d)
```

```
LpUrban=cos(fc,loc,tower,user,d)
LpSuburban=cos(fc,loc,tower,user,d)
LpOpen=cos(fc,loc,tower,user,d)
```

```
print(LpUrban,LpSuburban,LpOpen)
```

```
plt.figure()
plt.plot(d,LpUrban)
plt.plot(d,LpSuburban)
plt.plot(d,LpOpen)
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```

```
d_sort=d[np.argsort(d)]
Urban_sort=LpUrban[np.argsort(d)]
d_sort=d[np.argsort(d)]
Suburban_sort=LpSuburban[np.argsort(d)]
d_sort=d[np.argsort(d)]
Open_sort=LpOpen[np.argsort(d)]
```

```
plt.figure()
plt.plot(d_sort,Urban_sort)
plt.plot(d_sort,Suburban_sort)
plt.plot(d_sort,Open_sort)
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```
